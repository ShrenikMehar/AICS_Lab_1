```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from google.colab import drive 
drive.mount('/content/gdrive')
```

```python
data=pd.read_csv("gdrive/My Drive/AICS Datasets/Mall_Customers.csv")
data.head()
```

```python
df = data[["CustomerID", "Genre", "Age", "Annual Income (k$)", "Spending Score (1-100)"]]
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]
X.head()
```

```python
plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', data = X, s = 60)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```

```python
n = int(input("How many clusters do you want? "))
km=KMeans(n_clusters=n)
km.fit(X)
```

```python
y=km.predict(X)
y
```

```python
df["label"] = y
df.head()
```

```python
#Scatterplot of the clusters
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', hue = "label", legend = 'full', data = df, s = 60)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```

**If There are only 2 clusters**

```python
km=KMeans(n_clusters=2)
km.fit(X)
y=km.predict(X)
y
df["label"] = y
df.head()
#Scatterplot of the clusters
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', hue = "label", legend = 'full', data = df, s = 60)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```

**If There are 10 clusters**

```python
km=KMeans(n_clusters=10)
km.fit(X)
y=km.predict(X)
y
df["label"] = y
df.head()
#Scatterplot of the clusters
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'Annual Income (k$)', y = 'Spending Score (1-100)', hue = "label", legend = 'full', data = df, s = 60)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)') 
plt.title('Spending Score (1-100) vs Annual Income (k$)')
plt.show()
```
